### 10.2 데이터 합치기

- 데이터베이스 스타일로 DataFrame 합치기

- inner join

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
df2 = pd.DataFrame({'key': ['a', 'b', 'd'],
                    'data2': range(3)})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on='key',how='inner') # how='inner' 가 기본값
```

```python
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'], 'data1': range(7)})
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'], 'data2': range(3)})
```

```python
df3
```

```python
df4
```

```python
pd.merge(df3,df4,left_on='lkey',right_on='rkey')
```

- outer join

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
df2 = pd.DataFrame({'key': ['a', 'b', 'd'],
                    'data2': range(3)})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,how='left')
```

```python
pd.merge(df1,df2,how='right')
```

```python
pd.merge(df1,df2,how='outer') #양쪽에 있는거 다가져오기
```

- 색인 병합하기

```python
left1 = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'], 'value': range(6)})
right1 = pd.DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
```

```python
left1
```

```python
right1
```

```python
pd.merge(left1,right1,left_on='key',right_index=True)
```

```python
left2 = pd.DataFrame([[1., 2.], [3., 4.], [5., 6.]],
                     index=['a', 'c', 'e'],
                     columns=['Ohio', 'Nevada'])
right2 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],
                      index=['b', 'c', 'd', 'e'],
                      columns=['Missouri', 'Alabama'])
```

```python
left2
```

```python
right2
```

```python
pd.merge(left2,right2,left_index=True,right_index=True)
```

```python
left2.join(right2,how='inner') # join 함수는 인덱스 값을 기준으로 조인을 함 위와 같음 대신 default 값이 how='left'로 됨
```

```python
left1
```

```python
right1
```

```python
left1.join(right1,on='key',how='inner')   #pd.merge(left1,right1,left_on='key',right_index=True)
```

```python
df1
```

```python
df2
```

```python
pd.concat([df2,df1],axis=0)
```

```python
# 인덱스를 재 설정
```

```python
pd.concat([df1,df2],axis=0,ignore_index=True)
```

```python
pd.concat([df2,df1],axis=1)
```

```python
# 두 데이터 프레임에 공통된 행만 추출 (join='inner')
```

- 데이터베이스 스타일로 DataFrame 합치기

```python
df1 = pd.read_excel('./datasets/stock price.xlsx',index_col='id')
df2 = pd.read_excel('./datasets/stock valuation.xlsx',index_col='id')
```

```python
df1
```

```python
df2
```

```python
#id 를 조인 조건으로 해서 inner 조인
```

```python
pd.merge(df1,df2,on='id',how='inner') # how='inner' 가 기본값
```

```python
#id 를 조인 조건으로 해서 outer 조인
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on='id',how='outer') # how='inner' 가 기본값
```

```python
# 왼쪽 데이터프레임(df1)에서는 stock_name, 오른쪽데이터프레임(df2)에서는 name을 조인조건으로 하되
#left 조인
```

```python
pd.merge(df1,df2,left_on='stock_name',right_on='name',how='left') # how='inner' 가 기본값
```

```python
# 왼쪽 데이터프레임(df1)에서는 stock_name, 오른쪽데이터프레임(df2)에서는 name을 조인조건으로 하되
#right 조인
```

```python
pd.merge(df1,df2,left_on='stock_name',right_on='name',how='right') # how='inner' 가 기본값
```

- 색인 병합으로 데이터프레임 합치기

```python
pd.merge(df1,df2,left_index=True,right_index=True)
```

```python
df1.join(df2,how='inner')
```

```python
df1.join(df2,how='left')
```

```python
df1.join(df2,how='right')
```
