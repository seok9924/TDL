1. 데이터 분석 개요

## Step 1: 질문하기 (Ask questions)
데이터가 주어진 상태에서 질문을 할 수도 있고, 질문에 답할 수 있는 데이터를 수집할 수도 있다.

## Step 2: 데이터 랭글링 (Wrangle data)
데이터 랭글링 : 원자료(raw data)를 보다 쉽게 접근하고 분석할 수 있도록 데이터를 정리하고 통합하는 과정 (참고. 위키피디아)
세부적으로는 데이터의 수집(gather), 평가(assess), 정제(clean) 작업으로 나눌 수 있다.


## Step 3: 데이터 탐색 (Exploratory Data Analysis)
데이터의 패턴을 찾고, 관계를 시각화 하는 작업을 통해 데이터에 대한 직관을 극대화 한다.

## Step 4: 결론 도출 또는 예측 (Draw conclusions or make predictions)
- Step 3에서 분석한 내용을 근거로 질문에 대한 답과 결론을 도출 할 수 있다.
- 머신러닝 또는 통계 추정 과정을 거치게 되면 예측을 만들어 낼 수도 있다.

## Step 5: 결과 공유 (Communicate the results)
보고서, 이메일, 블로그 등 다양한 방법을 통해 발견한 통찰들을 공유할 수 있다.

# 2. Case Study

## Bike Sharing Demand

- 도시 자전거 공유 시스템 사용 예측
- [캐글](https://www.kaggle.com)의 [Bike Sharing Demand](https://www.kaggle.com/c/bike-sharing-demand)에서 `train.csv`와 `test.csv`를 다운로드
- 두 파일을 각각 datasets 디렉토리에 bike_train.csv bike_test.csv로 저장 

**datetime** : hourly date + timestamp  
**season** : 1 = 봄, 2 = 여름, 3 = 가을, 4 = 겨울  
**holiday**: 1 = 토, 일요일의 주말을 제외한 국경일 등의 휴일, 0 = 휴일이 아닌 날  
**workingday**: 1 = 토, 일요일의 주말 및 휴일이 아닌 주중, 0 = 주말 및 휴일  
**weather**:  
- 1 = 맑음, 약간 구름 낀 흐림  
- 2 = 안개, 안개 + 흐림  
- 3 = 가벼운 눈, 가벼운 비 + 천둥  
- 4 = 심한 눈/비, 천둥/번개  

**temp**: 온도(섭씨)   
**atemp**: 체감온도(섭씨)  
**humidity**: 상대습도  
**windspeed**: 풍속  
**casual**: 사전에 등록되지 않는 사용자가 대여한 횟수  
**registered**: 사전에 등록된 사용자가 대여한 횟수  
**count**: 대여 횟수

## Step 2: 데이터 랭글링 (Wrangle data)

- 데이터 적재

```python
import pandas as pd
```

데이터프레임=pd.read_csv(파일명)

```python
bike=pd.read_csv('./datasets/bike_train.csv')
```

```python
type(bike)
```

- 데이터 평가 

```python
bike.head() #앞의 5 열 정도를 훑어 볼 수있음 
```

```python
bike.info() # 칼럼의 데이터 타입 , 누락 건수 , 몇 개의 컬럼, 몇개의 샘플 
```

- 데이터 정제(누락된 값 처리, 잘못된 데이터 타입)

```python
#bike.datetime
bike['datetime']
```

```python
bike['datetime']=bike['datetime'].apply(pd.to_datetime)
```

```python
bike['year']=bike['datetime'].apply(lambda x: x.year ) # 이 함수에 의해 모든 데이터가 적용이 됨 
bike['month']=bike['datetime'].apply(lambda x: x.month ) # 이 함수에 의해 모든 데이터가 적용이 됨 
bike['hour']=bike['datetime'].apply(lambda x: x.hour ) # 이 함수에 의해 모든 데이터가 적용이 됨 
bike['dayofweek']=bike['datetime'].apply(lambda x: x.dayofweek ) # 이 함수에 의해 모든 데이터가 적용이 됨 
```

```python
bike.head()
```

```python
bike.info()
```

## Step3: 데이터 탐색(Exploratory Data Analysis)

- 기상정보(온도,체감온도,풍속,습도) 와 자전거 대여량의 관계

- 산점도로 확인 
- 상관계수로 확인

```python
bike.plot(kind='scatter',x='temp',y='count',alpha=0.3)
```

```python
import matplotlib.pyplot as plt
```

```python
fig,axes= plt.subplots(nrows=2,ncols=2,figsize=(14,8))
axes[0][0].scatter(bike['temp'],bike['count'],alpha=0.3)
axes[0][1].scatter(bike['atemp'],bike['count'],alpha=0.3)
axes[1][0].scatter(bike['windspeed'],bike['count'],alpha=0.3)
axes[1][1].scatter(bike['humidity'],bike['count'],alpha=0.3)
```

- 상관계수 (산점도를 통해 나온 결과를 수치화된 데잉터로 알려줌)

```python
bike.corr()
```

**분석**
- 질문 1에 대한 분석) 기상 정보 중 온도와 체감온도가 자전거 대여량에 영향을 미칠 것으로 보임 (산점도, 상관계수)

- **질문**2에 대한 분석)
- year, month, hour , dayof,week 특성에 따른 자전거 대여량이 어떻게 달라지는지 분석 
Monday=0,
- year, month, hour , dayof,week  : 범주형 데이터 끊겨 있는 데이터 
- count(자전거 대여량) : 수치형 데이터가 어떻게 달라지는지 파악할 때 막대그래프   barplot 유용

```python
import seaborn as sns 
```

```python
fig,axes= plt.subplots(nrows=2,ncols=2,figsize=(14,8))
sns.barplot(data=bike,x='year',y='count',ax=axes[0][0])
sns.barplot(data=bike,x='month',y='count',ax=axes[0][1])
sns.barplot(data=bike,x='hour',y='count',ax=axes[1][0])
sns.barplot(data=bike,x='dayofweek',y='count',ax=axes[1][1])
```

**분석결과**
- 연도별 평균 대여량은 2011년도보다 2012년도에 더많음
- 월별 평균 대여량은 6월이 가장많고 , 7~10월에도 많음 , 1월에 가장적음
- 시간대별 평균 대여량은 오전 8시 전후와 오후 5~6시 부근에 많음
