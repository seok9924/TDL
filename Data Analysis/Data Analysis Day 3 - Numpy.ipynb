### 3.2 배열 연산과 조건절 표현하기

```python
x_list=[1,2,3,4]
y_list=[5,6,7,8]
cond_list=[True,True,False,False]
result=[]
for x,y,c in zip (x_list,y_list,cond_list) :
    if c:
        result.append(x)
    else:
        result.append(y)
result
```

```python
x_arr=np.array(x_list)
y_arr=np.array(y_list)
cond_arr=np.array(cond_list)
result=np.where(cond_arr, x_arr, y_arr )
result

```

계단 오르 내리기 
- 동전을 던졌을 때 앞이 나오면 계단을 1칸 올라가기 
- 동전을 던졌을 때 뒤가 나오면 계단 1칸 내려가기 
- 현재 위치 position 0 에서 시작 
- 1000번 동전을 던졌을 때의 위치의 히스토리 : walk 리스트에

```python
import random
```

```python
position = 0 # 계단 1칸 오르면  +1 계단 1칸 내려가게 되면 -1 
steps =1000

walk=[position]
step=0
random.randint(0,1)
for i in range(steps):
    draw=random.randint(0,1)
    if draw==1:
        step=1
    else:
        step=-1
    position+=step
    walk.append(position)
plt.plot(walk)
```
