## 1. 정규 방정식을 사용한 선형회귀

- y=3X + 4

```python
X= 2 *np.random.rand(100,1) #100 x1 의 데이터 
y= 3*X+4+np.random.randn(100,1)
```

```python
plt.plot(X,y,'.')
```

```python
#np.ones((100,1))
```

```python
X_b=np.c_[np.ones((100,1)),X]   # 세타값 붙여주기 
```

```python
Image('./images/img1.png')
```

- 정규방정식 공식을 np.linalg 함수를 이용해서 해를 구함

```python
theta_best=np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
theta_best #앞이 세타제로 뒤가 세타 원 theta 0: 절편 theta1: 기울기
```

- scikit-learn 제공 LinearRegression 사용

- 배치 경사하강법

```python
m= 100
theta= np.random.randn(2,1)
eta=0.1

n_iterations=1000
for iteration in range(n_iterations) :
    gradients= 2/m *X_b.T.dot(X_b.dot(theta)-y) #gradients ()지만 thata1에 대한 미분, theta2 에 대한 미분 
    theta= theta- eta*gradients #theta(theta0, theta 1 포함)
```

```python
theta  # y= 3X+4
```

- 확률적 경사하강법

```python
np.random.seed(42)
```

```python
m= 100
theta= np.random.randn(2,1)
n_epochs=50
t0,t1=5,50
def learning_schedule(t):
    return t0/(t+t1) 

for epoch in range(n_epochs) : #n_epochs epoch은 모든 데이터가 학습에 참여했을때 1epoch 이라고 함 
    for i in range(m) : # m : 100개의 샘플수 (100개의 샘플중에 여러번 뽑히는 샘플도 있고 전혀 선택되지 않는 애도 있고)
        random_index=np.random.randint(m)
        xi = X_b[random_index:random_index+1]  # 1이 아니라 10이면 확장
        yi = y[random_index:random_index+1]
        gradients= 2 *xi.T.dot(xi.dot(theta)-yi) #gradients ()지만 thata1에 대한 미분, theta2 에 대한 미분 
        eta=learning_schedule(epoch*m+i)  #step수가 증가함에 따라 eta(학습률이) 작아지게 하는 효과
        theta= theta- eta*gradients #theta(theta0, theta 1 포함)
#         mse=mean((실제값들-예측값들)**2)
#         print(mse)
```

## 4. 규제 모델

### 4.1 릿지 회귀 - L2 규제

```python
np.random.seed(42)
m=20
X=3*np.random.rand(m,1)
y=0.5*X+1+np.random.randn(m,1)/1.5
plt.plot(X,y,'.')
```

```python
#선형회귀 모델
lin_reg=LinearRegression()
lin_reg.fit(X,y)
```

```python
lin_reg.intercept_,lin_reg.coef_
```

```python
#릿지회귀 모델(L2 규제) -해석적으로 해를 구함
from sklearn.linear_model import Ridge
ridge_reg=Ridge()
ridge_reg.fit(X,y)
```

```python
ridge_reg.intercept_,ridge_reg.coef_  #theta 1이 규제가 됨 상대적으로 기본보다 값이 줄어듬
#절편 theta0는 규제의 범위에 포함되지 않음 
```

```python
#릿지회귀 모델(L2 규제) -경사하강법으로 해를 구함
from sklearn.linear_model import SGDRegressor

# 규제 없이
sgd_reg=SGDRegressor(penalty=None,random_state=42)
sgd_reg.fit(X,y.ravel())
print(sgd_reg.intercept_,sgd_reg.coef_)

#규제 추가
sgd_reg_l2=SGDRegressor(penalty='l2',alpha=0.1,random_state=42)
sgd_reg_l2.fit(X,y.ravel())
print(sgd_reg_l2.intercept_,sgd_reg_l2.coef_)



```

### 4.2 라쏘 회귀 - L1 규제

```python
from sklearn.linear_model import Lasso

lasso_reg=Lasso(alpha=0.1,random_state=42)
lasso_reg.fit(X,y)
lasso_reg.intercept_,lasso_reg.coef_
```

```python
# 규제 없이
sgd_reg=SGDRegressor(penalty=None,random_state=42)
sgd_reg.fit(X,y.ravel())
print(sgd_reg.intercept_,sgd_reg.coef_)

#규제 추가
sgd_reg_l1=SGDRegressor(penalty='l1',alpha=0.1,random_state=42)
sgd_reg_l1.fit(X,y.ravel())
print(sgd_reg_l1.intercept_,sgd_reg_l1.coef_)
```

### 4.3  엘라스틱넷

```python
# 엘라스틱넷 (l1규제, l2 규제 )
from sklearn.linear_model import ElasticNet

elastic_reg=ElasticNet(alpha=0.1,l1_ratio=0.5,random_state=42)
elastic_reg.fit(X,y)
elastic_reg.intercept_,elastic_reg.coef_
```

```python
# 규제 없이
sgd_reg=SGDRegressor(penalty=None,random_state=42)
sgd_reg.fit(X,y.ravel())
print(sgd_reg.intercept_,sgd_reg.coef_)

#규제 추가
sgd_reg_l1=SGDRegressor(penalty='elasticnet',alpha=0.1,random_state=42)
sgd_reg_l1.fit(X,y.ravel())
print(sgd_reg_l1.intercept_,sgd_reg_l1.coef_)
```
