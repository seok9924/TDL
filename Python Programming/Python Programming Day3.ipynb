**알파벳별 빈도 사전 등록**
- 다음의 주어진 문자열 s에서 알파벳 별 빈도를 사전형인 d에 등록하는 파이썬 코드블럭을 작성하시오. 

```
s = ‘life is short, so python is easy.’ 
punct = ‘,. ’ 
d = {} 

결과 print(d) (순서는 상관 없음)
{'a': 1, 'e': 2, 'f': 1, 'i': 3, 'h': 2, 'l': 1, 'o': 3, 'n': 1, 'p': 1, 's': 5, 'r': 1, 't': 2, 'y': 2}


````python
s = "life is short, so python is easy."
alpha="abcdefghijklmnopqrstuvwxyz"
d={}
for a in alpha :
    if a in s :
        d[a]=s.count(a)
print(d)
````

````python
s = "life is short, so python is easy."
punct=',. '
d2={}
for c in s:
    if c in punct:
        continue
    if c not in d2:
        d2[c]=0 # 0으로 초기화 해주고 
    if c in d2:
        d2[c]+=1
print(d2)
    
````

FizzBuzz 문제

- 1에서 100까지 출력
- 3의 배수는 Fizz 출력
- 5의 배수는 Buzz 출력
- 3과 5의 공배수는 FizzBuzz 출력


````python
for i in range(1,101):
    
    if i%3==0 and i%5==0:
        print("FizzBuzz")
        
    elif i%3==0 :
        print("Fizz")
        
    elif i%5==0 :
        print("Buzz")
    else:
        print(i) 
````

````python

push=int(input())
for j in range(push):
    x=[]
    i=0
    while i<6 :
        k=random.randint(1,45)
       # take=radom.sample(range(1,46),6)
        if k not in x :
            x.append(k)
            i+=1
        else:
            continue
    x.sort()
    print(x)
    
````
